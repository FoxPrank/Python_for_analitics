```python
import pandas as pd
```

```python
df = pd.read_csv("winequality-red.csv", sep = ";")
df.head()
```

```python
df['good'] = df['quality'].apply(lambda x: 1 if x > 5 else 0)
df.sample(n=15)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df.drop(columns = ['quality'],axis = 1,inplace=True)
df.head()
```

```python
correlation = df.corr()
correlation
```

Анализ влияния химиеских каратктеристик красного вина на оценку качества продукта.

Сначала выявим факторы, которые в большей степени коррелируют с оценкой выше 5

```python
plt.figure(figsize = (10,7))
sns.heatmap(correlation, cmap = 'Blues', annot = True)
plt.title('Тепловая карта корреляции')
plt.show()
```

Факторы с наибольшей прямой корреляцией:
1. alcohol	
2. sulphates
3. citric acid

Факторы с наибольшей обратной корреляцией:
1. volatile acidity
2. total sulfur dioxide
3. density

```python
df_full = pd.read_csv("winequality-red.csv", sep = ";")
df_full.head()
```

```python
df_full['good'] = df_full['quality'].apply(lambda x: 1 if x > 5 else 0)
df_full.sample(n=15)
```

```python
data = df_full.groupby('quality')[['alcohol','sulphates','citric acid']].mean().reset_index()

```

```python
plt.figure(figsize = (10,7))

plt.plot(data['quality'], data['alcohol'])
plt.plot(data['quality'], data['sulphates'])
plt.plot(data['quality'], data['citric acid'])

plt.title('График зависимости')
plt.xlabel('Оценка качества') 
plt.ylabel('Среднее значение показателя')
plt.legend(['alcohol','sulphates','citric acid'])

plt.show()
```

Таким образом, детальный анализ зависимости топовых по прямой корреляции факторов подтверждает вывод, что оценка возрастает при увеличении среднего значения по показателям 'alcohol','sulphates','citric acid'

Посмотрим распределение хороших оценках на показателях с наибольшей отрицательной корреляцияей

```python
data1 = df_full.groupby('good')[['volatile acidity', 'total sulfur dioxide','density']].mean().unstack('good').unstack('good').reset_index()
data1.rename(columns = {'index' : 'factor'}, inplace = True) 
data1.columns.name = None
data1.head()
```

```python
import numpy as np
```

```python
plt.figure(figsize = (10,7))

n_ticks = np.arange(len(data1[0]))

offset = 0.2
w = 0.4

plt.bar(n_ticks - offset, data1[0], width=w)
plt.bar(n_ticks + offset, data1[1], width=w)

plt.title('сравнение средних значений фактора плохих и хороших вин')
plt.xlabel('factor')
plt.ylabel('Среднее')
plt.legend(['Плохие','Хорошие'])
plt.xticks(n_ticks, data1['factor'], rotation = 30) 
plt.show()
```

Очевидно, что у хорошего вина среднее зачение total sulfur dioxide должно быть ниже.
Подтвердим данный вывод, построив совеместное распределение по переменной total sulfur dioxide и quality

```python
sns.jointplot(x=df_full['quality'], y=df_full['total sulfur dioxide'], data=df_full, kind='reg')
```

```python
data2 = df_full['good'].value_counts()
data2.head()
```

```python
plt.figure(figsize = (10,7))

plt.pie(data2, autopct='%1.1f%%')

plt.title('Доля хороших (1) и плохих (0) оценок во всём исследовании')
plt.legend(data2.index)

plt.show()
```

```python
В представленных данных хороших оценок чуть больше, чем плохих, это означает, что выборка достаточна разнообразна 
и выводы о влиянии факторов можно считать достоверными (без перекоса статистической выборки).
```
